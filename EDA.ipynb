```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Spotify_Youtube.csv")
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.head()
```

```python
#Change polularity metrics into millions
df['Views'] = df['Views'].divide(1000000)
df['Likes'] = df['Likes'].divide(1000000)
df['Stream'] = df['Stream'].divide(1000000)
```

```python
#Album vs single
df_pull1 = df[['Artist', 'Album_type', 'Track', 'Views', 'Likes', 'Stream']]
```

```python
df_pull1.head()
```

```python
df_album = df_pull1[df_pull1['Album_type'].str.contains('album')]
df_single = df_pull1[df_pull1['Album_type'].str.contains('single')]
```

```python
df_album.shape
```

```python
df_single.shape
```

```python
pie_data = [14926, 5004]
pie_labels = ['Album', 'Single']
pie_colors = sns.color_palette('pastel')[0:2]
plt.pie(pie_data, labels = pie_labels, colors = pie_colors, autopct='%.0f%%')
plt.show()
```

```python
#Top 10 album
top10_album = df_album.sort_values(by='Views', ascending=False).head(10)
top10_album
```

```python
#Top 10 Single
top10_single = df_single.sort_values(by='Views', ascending=False).head(10)
top10_single
```

```python
df_album.describe()
```

```python
df_single.describe()
```

```python
#Duration vs poularity 
df_plot = df[['Duration_ms', 'Likes']].sort_values(by='Duration_ms')
df_plot['Duration_ms'] = df_plot['Duration_ms'].divide(1000*60)
df_plot
```

```python
#Duration vs Likes
df_plot.plot(kind = 'line', x = 'Duration_ms', y = 'Likes')

plt.show()
```

```python
#Duration vs Stream
df.plot(kind = 'scatter', x = 'Duration_ms', y = 'Stream')

plt.show()
```

```python
#Duration summary
df_mins = df['Duration_ms'].divide(1000*60)
df_mins.describe()
```

```python
#Duration density plot
sns.displot(df_mins)
```
